# Importing libraries and reading data

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 12]

df = pd.read_csv('https://github.com/terezaif/workshops_data_exploration/raw/master/datasets/yellow_tripdata_2016-06_sample30.csv.zip', compression='zip')
df.head()
```

# Q1 - How are the trip_distance and tip_amount related?

```python
df.plot.scatter('trip_distance', 'tip_amount');
```

# Q2 - What is the correlation between the variables above? What kind of information can you extract from that?

```python
df[['trip_distance', 'tip_amount']].corr()
```

# Q3 - How is the fare distribution between vendor_id = [1, 2]? Anything suspicious with this data?

```python
df.boxplot('fare_amount', by='VendorID');
```
